```python
import requests
from bs4 import BeautifulSoup, Comment

def gets_soup(url):
    """Fetches and parses HTML content from a given URL into a BeautifulSoup object."""
    response = requests.get(url)
    html = response.content
    return BeautifulSoup(html, 'html.parser')

def clean_html(soup):
    """Cleans the BeautifulSoup object by removing unwanted tags, attributes, and comments."""
    # Remove the head tag and its contents
    if soup.head:
        soup.head.decompose()
    
    # Remove script and style elements
    for script_or_style in soup.find_all(["script", "style"]):
        script_or_style.decompose()
    
    # Remove comments
    comments = soup.find_all(string=lambda text: isinstance(text, Comment))
    for comment in comments:
        comment.extract()
    
    # Define unwanted attributes including event handlers
    unwanted_attributes = ["class", "id", "style", "onclick", "onerror", "onload"]
    
    # Remove attributes by pattern (e.g., data-*, aria-*)
    for tag in soup.find_all():
        for attribute in list(tag.attrs):
            if attribute.startswith(("data-", "aria-", "on")) or attribute in unwanted_attributes:
                del tag.attrs[attribute]
    
    # Remove unwanted tags and additional layout-oriented or non-content tags
    unwanted_tags = ['header', 'footer', 'nav', 'aside', 'svg', 'form', 'input', 'textarea', 'button', 'select']
    for tag in unwanted_tags:
        for element in soup.find_all(tag):
            element.decompose()
    
    # Unwrap certain tags to keep their content but remove the tag itself
    for tag_name in ['span', 'div', 'section']:
        for tag in soup.find_all(tag_name):
            tag.unwrap()
    
    return soup.prettify()

def count_tokens(cleaned_html):
    """Counts the number of whitespace-separated tokens in the cleaned HTML content."""
    tokens = cleaned_html.split()
    return len(tokens)

# Usage
url = 'https://www.gifthampersuk.co.uk/chocolate-mountain-gift-hamper-uk'
soup = gets_soup(url)
cleaned_html = clean_html(soup)
token_count = count_tokens(cleaned_html)
print(f"Number of tokens: {token_count}")
```
